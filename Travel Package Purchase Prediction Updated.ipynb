```python
# Z SCORE
outliers = []
def detect_outliers_zscore(data):
    thres = 3
    mean = np.mean(data)
    std = np.std(data)
    print("Mean:",format(mean,".2f") , "Std:",format(std,".2f"))
    
    for i in data:
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers# Driver code

df_income_outliers = df["MonthlyIncome"]
sample_outliers = detect_outliers_zscore(df_income_outliers)
print("Outliers from Z-scores method: ", sample_outliers)
```

```python
#Z-score(Removing outlier using Z-score)

def remove_outliers_zscore(df, column, threshold=3):
    # Calculate the Z-score for the given column
    z_scores = np.abs((df[column] - df[column].mean()) / df[column].std())

    # Filter the DataFrame to remove outliers
    filtered_df = df[z_scores < threshold]

    return filtered_df

# Remove outliers using Z-score
filtered_df = remove_outliers_zscore(df, 'MonthlyIncome')

filtered_df.describe()
```

```python
filtered_df[filtered_df["ProdTaken"] == 1].count()
```

```python
#Removing outliers using IQR(Inter quartile range)

def remove_outliers_iqr(df, column):
    # Calculate the first quartile (Q1) and third quartile (Q3)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1

    # Calculate the lower and upper limits for outliers
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Filter the DataFrame to remove outliers
    filtered_df = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]

    return filtered_df


# Remove outliers using IQR
filtered_df1 = remove_outliers_iqr(df, 'MonthlyIncome')
filtered_df1.describe()

```

###  Defining Models

```python
models={
    "Logistic_Regression":LogisticRegression(),
    "Random_Forest_Classifier":RandomForestClassifier(),
    "Naive_Bayes_Classifier":GaussianNB(),
    "Decision_Tree_Classifier":DecisionTreeClassifier(),
    "Support_Vector_Machine":SVC(),
    "KNeighborsClassifier":KNeighborsClassifier()
}
```

```python
list(models.values())[:]
```

```python
def Model_Training(model_dict,x_train_data,y_train_data):
    acc_report={}
    for i in range(len(list(model_dict))):
        model=list(model_dict.values())[i]
        model.fit(x_train_data,y_train_data)
        y_pred_train=model.predict(x_train_data)
        report=classification_report(y_pred_train,y_train_data)
        accuracy=accuracy_score(y_pred_train,y_train_data)
        print(f"Model:{model}")
        print(f"accuracy for training data for {model}: {accuracy}")
        print(report)
        acc_report[model]=accuracy
    return acc_report
```

```python
report=Model_Training(models,x_train,y_train)
```

```python
report
```
